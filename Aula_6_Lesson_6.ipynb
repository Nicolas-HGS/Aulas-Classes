```python
import cv2
import matplotlib.pyplot as plt

figure = cv2.imread("original.jpeg")

figure = cv2.cvtColor(figure, cv2.COLOR_BGR2RGB)

plt.imshow(figure)
plt.show()
```

```python
a = figure[0,0]

b = figure[250,0]

print(a)

print(bin(a[0]),bin(a[1]),bin(a[2]))

print(b)

print(bin(b[0]),bin(b[1]),bin(b[2]))

xor_operation = a ^ b

print(bin(xor_operation[0]),bin(xor_operation[1]),bin(xor_operation[2]))

print(xor_operation)

print(cv2.bitwise_xor(figure[0,0],figure[250,0]))
```

```python
fig, axs = plt.subplots(1, 3, layout='constrained', figsize = [10,10])

axs[0].imshow(figure)

axs[0].set_title('Imagem original / Original Image')

axs[0].set_axis_off()

figure_2 = cv2.imread("modified.jpeg")

figure_2 = cv2.cvtColor(figure_2, cv2.COLOR_BGR2RGB)

axs[1].imshow(figure_2)

axs[1].set_title('Imagem modificada / Modified image')

axs[1].set_axis_off()

figure_3 = figure ^ figure_2

axs[2].imshow(figure_3)

axs[2].set_title('Imagem resultante / Resulting Image')

axs[2].set_axis_off()

plt.show()
```

```python
img_result = cv2.bitwise_xor(figure,figure_2)

fig, axs = plt.subplots(1, 2, layout='constrained', figsize = [13,13])

axs[0].imshow(img_result)

axs[0].set_title('Operação XOR pela função da biblioteca / XOR operation by library function')

axs[1].imshow(figure_3)

axs[1].set_title('Operação XOR pelo operador ^ / XOR operation by the ^ operator')

axs[0].set_axis_off()
axs[1].set_axis_off()

plt.show()
```
