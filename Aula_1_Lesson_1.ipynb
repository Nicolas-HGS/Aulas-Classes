```python
import cv2
```

```python
figure = cv2.imread('colorido.jpg')
```

```python
print(figure)
```

```python
print(figure.shape)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(figure)
```

```python
figure = cv2.cvtColor(figure, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(figure)
plt.show()
```

```python
figure_red = figure.copy()

figure_red[:,:,1:] = 0

print("Red layer of the image:\n")
print(figure_red,"\n")

print("Original Image:\n")
print(figure)
```

```python
plt.imshow(figure_red)
plt.show()
```

```python
figure_green = figure.copy()

figure_green[:,:,0] = 0

print("Green layer of the image:\n")
print(figure_green,"\n")

print("Original Image:\n")
print(figure)
```

```python
plt.imshow(figure_green)
plt.show()
```

```python
figure_blue = figure.copy()

figure_blue[:,:,0] = 0
figure_blue[:,:,1] = 0

print("Blue layer of the image:\n")
print(figure_blue,"\n")

print("Original Image:\n")
print(figure)
```

```python
plt.imshow(figure_blue)
plt.show()
```
