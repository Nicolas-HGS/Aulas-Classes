```python
import cv2
import matplotlib.pyplot as plt

figure = cv2.imread("original.jpeg")

figure = cv2.cvtColor(figure, cv2.COLOR_BGR2RGB)

plt.imshow(figure)
plt.show()
```

```python
pixel_gray = figure[600,590]

print(pixel_gray)

intensity_red = int(figure[600,590][0])

intensity_green = int(figure[600,590][1])

intensity_blue = int(figure[600,590][2])

print(intensity_red,intensity_green,intensity_blue)

average_intensities = ((intensity_red+intensity_green+intensity_blue)//3)

print(average_intensities)

pixel_gray[:] = average_intensities

print(pixel_gray)
```

```python
fig,axs = plt.subplots(1, 3, layout='constrained', figsize = [10,10])

axs[0].imshow(figure)

axs[0].set_title('Imagem original / Original Image')

axs[0].set_axis_off()

figure_gray = figure.copy()

height = figure_gray.shape[0]

width = figure_gray.shape[1]

for i in range(height):
  for j in range(width):

    intensity_red = int(figure_gray[i,j][0])

    intensity_green = int(figure_gray[i,j][1])

    intensity_blue = int(figure_gray[i,j][2])

    figure_gray[i,j] = ((intensity_red+intensity_green+intensity_blue)//3)

axs[1].imshow(figure_gray)

axs[1].set_title('Resultado do código / Code Result')

axs[1].set_axis_off()

figure_gray_func = cv2.cvtColor(figure, cv2.COLOR_RGB2GRAY)

axs[2].imshow(figure_gray_func, cmap='gray')

axs[2].set_title("Função da biblioteca / Library function")

axs[2].set_axis_off()

plt.show()
```
