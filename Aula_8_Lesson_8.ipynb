```python
import cv2
import matplotlib.pyplot as plt

figure = cv2.imread("original.jpeg")

figure = cv2.cvtColor(figure, cv2.COLOR_BGR2RGB)

plt.imshow(figure)
plt.show()
```

```python
import numpy as np

a = figure[207,300]

b = figure[207,353]

print(a)

print(b)

a = np.array([a])

b = np.array([b])

print(a-b)

print(cv2.subtract(a,b))
```

```python
fig,axs = plt.subplots(3, 3, layout='constrained', figsize = [10,10])

axs[0,0].imshow(figure)

axs[0,0].set_title('Imagem original / Original Image')

axs[0,0].set_axis_off()

figure_2 = cv2.imread("mask-1.jpeg")

figure_2 = cv2.cvtColor(figure_2, cv2.COLOR_BGR2RGB)

axs[0,1].imshow(figure_2)

axs[0,1].set_title('Máscara / Mask')

axs[0,1].set_axis_off()

figure_3 = figure - figure_2

axs[0,2].imshow(figure_3)

axs[0,2].set_title('Imagem resultante / Resulting Image')

axs[0,2].set_axis_off()

figure_4 = cv2.imread("mask-2.jpg")
figure_4 = cv2.cvtColor(figure_4, cv2.COLOR_BGR2RGB)
figure_5 = cv2.imread("mask-3.jpg")
figure_5 = cv2.cvtColor(figure_5, cv2.COLOR_BGR2RGB)

axs[1,0].imshow(figure)
axs[1,0].set_axis_off()
axs[1,1].imshow(figure_4)
axs[1,1].set_axis_off()
axs[1,2].imshow(figure - figure_4)
axs[1,2].set_axis_off()
axs[2,0].imshow(figure)
axs[2,0].set_axis_off()
axs[2,1].imshow(figure_5)
axs[2,1].set_axis_off()
axs[2,2].imshow(figure - figure_5)
axs[2,2].set_axis_off()

plt.show()

```
