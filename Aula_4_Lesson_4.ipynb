```python
import cv2
import matplotlib.pyplot as plt

figure = cv2.imread("imagem.jpg")

figure = cv2.cvtColor(figure, cv2.COLOR_BGR2RGB)

plt.imshow(figure)
plt.show()
```

```python
a = figure[0,0]

b = figure[250,0]

print(a)

print(bin(a[0]),bin(a[1]),bin(a[2]))

print(b)

print(bin(b[0]),bin(b[1]),bin(b[2]))

or_operation = a | b

print(bin(or_operation[0]),bin(or_operation[1]),bin(or_operation[2]))

print(or_operation)

print(cv2.bitwise_or(figure[0,0],figure[250,0]))
```

```python
fig = plt.figure(figsize=[10,10])

ax1 = fig.add_subplot(1, 3, 1)

plt.title("Imagem original / Original Image")

figure_2 = cv2.imread("mask.jpg")

ax2 = fig.add_subplot(1, 3, 2)

plt.title("Máscara / Mask")

figure_3 = figure | figure_2

ax3 = fig.add_subplot(1, 3, 3)

plt.title("Imagem resultante / Resulting Image")

ax1.imshow(figure)

ax1.set_axis_off()

ax2.imshow(figure_2)

ax2.set_axis_off()

ax3.imshow(figure_3)

ax3.set_axis_off()

plt.show()

```

```python
img_result = cv2.bitwise_or(figure,figure_2)

fig = plt.figure(figsize=[10,10])

ax1 = fig.add_subplot(1, 2, 1)

ax2 = fig.add_subplot(1, 2, 2)

ax1.imshow(img_result)

ax2.imshow(figure_3)

ax1.set_axis_off()
ax2.set_axis_off()

plt.show()
```
