```python
import cv2

import matplotlib.pyplot as plt

figure = cv2.imread("imagem.jpg")

figure = cv2.cvtColor(figure, cv2.COLOR_BGR2RGB)

plt.imshow(figure)
plt.show()
```

```python
a = figure[0,0]

print(a)

print(bin(a[0]),bin(a[1]),bin(a[2]))

not_operation = ~a

print(bin(not_operation[0]),bin(not_operation[1]),bin(not_operation[2]))

print(not_operation)

print(cv2.bitwise_not(figure[0,0]))
```

```python
fig = plt.figure(figsize=[10,10])

ax1 = fig.add_subplot(1, 2, 1)

plt.title("Imagem original / Original Image")

figure_2 = ~figure

ax2 = fig.add_subplot(1, 2, 2)

plt.title("Imagem resultante / Resulting Image")

ax1.imshow(figure)

ax1.set_axis_off()

ax2.imshow(figure_2)

ax2.set_axis_off()

plt.show()
```

```python
img_result = cv2.bitwise_not(figure)

fig = plt.figure(figsize=[10,10])
ax1 = fig.add_subplot(1, 2, 1)
plt.title("Operador ~ / ~ Operator")
ax2 = fig.add_subplot(1, 2, 2)
plt.title("Função da biblioteca / Library Function")
ax1.imshow(figure_2)
ax2.imshow(img_result)
ax1.set_axis_off()
ax2.set_axis_off()

plt.show()
```
