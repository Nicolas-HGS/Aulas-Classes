```python
import cv2
import matplotlib.pyplot as plt

figure = cv2.imread("imagem.jpg")

figure = cv2.cvtColor(figure, cv2.COLOR_BGR2RGB)

plt.imshow(figure)
plt.show()
```

```python
manipulated_figure = figure.copy()

manipulated_figure[:,:,1:] = 0

plt.imshow(manipulated_figure)
plt.show()
```

```python
manipulated_figure[:,:,1:] = figure[:,:,1:]

plt.imshow(manipulated_figure)
plt.show()
```

```python
height = manipulated_figure.shape[0]

width = manipulated_figure.shape[1]

manipulated_figure[0:height//2,:] = figure[height//2:height,:]

manipulated_figure[height//2:height,:] = figure[0:height//2,:]

plt.imshow(manipulated_figure)
plt.show()
```

```python
manipulated_figure = figure.copy()

manipulated_figure[:,0:width//2] = figure[:,(width//2)+1:width]

manipulated_figure[:,(width//2)+1:width] = figure[:,0:width//2]

plt.imshow(manipulated_figure)
plt.show()
```

```python
manipulated_figure = figure.copy()

manipulated_figure[0:height//2,:,0] = 0
manipulated_figure[0:height//2,:,2] = 0

manipulated_figure[height//2:height,:,1] = 0
manipulated_figure[height//2:height,:,2] = 0

plt.imshow(manipulated_figure)
plt.show()
```

```python
manipulated_figure = cv2.cvtColor(manipulated_figure, cv2.COLOR_RGB2BGR)

cv2.imwrite("manipulated_figure.jpg", manipulated_figure)
```
