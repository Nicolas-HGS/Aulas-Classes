```python
import cv2
import matplotlib.pyplot as plt

figure = cv2.imread("imagem.jpg")

figure = cv2.cvtColor(figure, cv2.COLOR_BGR2RGB)

plt.imshow(figure)
plt.show()
```

```python
import numpy as np

a = figure[207,300]

b = figure[207,353]

print(a)

print(b)

a = np.array([a])

b = np.array([b])

print(a*b)

n_a = a.astype(int)
n_b = b.astype(int)

print(n_a*n_b)

print(cv2.multiply(a,b))

print(np.minimum(n_a*n_b, 255))
```

```python
fig,axs = plt.subplots(1, 4, layout='constrained', figsize = [13,13])

axs[0].imshow(figure)

axs[0].set_title('Imagem original / Original Image')

axs[0].set_axis_off()

mask = cv2.imread("mask - multiplication.jpg")

mask = cv2.cvtColor(mask, cv2.COLOR_BGR2RGB)

axs[1].imshow(mask)

axs[1].set_title('Máscara / Mask')

axs[1].set_axis_off()

figure_3 = figure.copy().astype(int)

figure_3 = np.minimum(figure_3*mask.astype(int),255)

axs[2].imshow(figure_3)

axs[2].set_title('Imagem resultante / Resulting Image')

axs[2].set_axis_off()

figure_4 = cv2.multiply(figure,mask)

axs[3].imshow(figure_4)

axs[3].set_title('Função da biblioteca / Library function')

axs[3].set_axis_off()

plt.show()
```

```python
subtraction = np.subtract(figure_3,figure_4)
module = np.abs(subtraction)
sum = np.sum(module)

print(sum)

q = np.array([[-5,4,3]])
j = np.array([[-1,0,3]])
sub = np.subtract(q,j)
print(sub)
mod = np.abs(sub)
print(mod)
print(np.sum(sub))
print(np.sum(mod))
```
